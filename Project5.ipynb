```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from a CSV file
data = pd.read_csv('international-tourist-arrivals-by-region-of-origin.csv')

# Filter the data for specific regions (e.g., Africa and Americas)
selected_regions = ["Africa", "Americas","East Asia and the Pacific","Europe","Middle East","South Asia","Not classified"]
filtered_data = data[data["Entity"].isin(selected_regions)]

# Pivot the data for plotting
pivot_data = filtered_data.pivot(index="Year", columns="Entity", values="International tourist arrivals by region")

# Create a line plot
pivot_data.plot(kind="line")
plt.title("International Tourist Arrivals by Region")
plt.xlabel("Year")
plt.ylabel("Arrivals")
plt.legend(title="Region")
plt.grid(True)

# Show the plot
plt.show()
```
